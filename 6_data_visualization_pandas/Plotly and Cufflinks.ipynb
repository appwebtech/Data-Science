```python
# Creating some data 
df = pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
```

```python
df2
```

```python
# Using Cufflinks is straightforward.
# If I call the plot fn, matplotlib will be called to to execute the plot. 

df.plot()
```

```python
# Changing it to iplot interactive image which I can zoom and interact with it. 
df.iplot()
```

```python
# Creating other plot types. 

# Scatter Plot 
df.iplot(kind='scatter',x='A',y='B',mode='markers')
```

```python
# Bar plots 

df2.iplot(kind='bar',x='Category',y='Values')
```

```python
# We can create aggregate functions for our plots. 
# If I had to create a plot for the whole df, it would show a bunch of bars 
# which dont make sense. 

df.iplot(kind='bar')
```

```python
# But I can narrow down and plot for instance the sum of each data frame. 

df.sum().iplot(kind='bar')
```

```python
# Box plots 

df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[500,400,300,200,100]})
```

```python
df3
```

```python
# Surface plot 
# We'll use the data above to create a 3D surface plot.

df3.iplot(kind='surface')
```

```python
# Poking things around, I'll change the 'z' co-ordinates to get a different surface 

df4 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df4.iplot(kind='surface')
```

```python
# Changing color scale 

df4.iplot(kind='surface',colorscale='rdylbu')
```

```python
# Histogram plot 
# You can specify the number of bins

df['A'].iplot(kind='hist',bins=50)
```

```python
# Careful! You'll get an overlapping plot if you pass the whole data frame.

# To see a particular plot you can turn them on and off. 

df.iplot(kind='hist')
```

```python
# Spread plot 
# Spread plots are used mostly to compare stock data. 

# I'll get a line plot with a sub-plot of a 'spead plot' against each other. 
df[['A','B']].iplot(kind='spread')
```

```python
# Bubble plot 

# Bubble plots are common in comparisons of GDP's against populations, etc

df.iplot(kind='bubble',x='A',y='B',size='C')
```
