## Pandas built-in data visualization



```python
import numpy as np
```

```python
import seaborn as sns
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
# Getting an histogram for all values on the 'A' column 

df1['A'].hist(bins=30)
```

```python
# We can change the style for a better looking graph if we incorporate seaborn
# Running the cell again and setting a 'darkgrid' 

df1['A'].hist(bins=30)
sns.set_style('darkgrid')
```

```python
# Another way to call a plot 

df1['A'].plot(kind='hist', bins=30)
```

```python
# You can also call hist directly to make a plot 

df1['A'].plot.hist(bins=30)
```

```python
# Plotting with the next data set. (df2)

df2.plot.area()
```

```python
# Transparency 

df2.plot.area(alpha=0.4)
```

```python
# Bar plots 
df2.plot.bar()
```

```python
# For a stacked option 

df2.plot.bar(stacked = True)
```

```python
df1['A'].plot.hist(bins=50)
```

```python
# Line plots 

# df1.plot.line(x=df1.index,y='B') (This plot throws errors)

df1['B'].plot(kind='line', y='B',x=df1.index) # I improvised this
```

```python
df1['B'].plot(kind='line', y='B',x=df1.index,figsize=(12,3),lw=1)
```

```python
# Scatter plots

df1.plot.scatter(x='A',y='B')
```

```python
# Setting colors for the scatter plot

df1.plot.scatter(x='A',y='B',c='C')
```

```python
# Box plot 

df2.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])
```

```python
df.head()
```

```python
# Hex plot 
# It's an hexagonal plot with bins which are small to be visible. 

df.plot.hexbin(x='a',y='b')
```

```python
# Changing bins visibility 

df.plot.hexbin(x='a',y='b',gridsize=25, cmap='coolwarm')
```

```python
# For a KDE plot

df2['a'].plot.kde()
```

```python
# Comparing KDE for the whole df 

df2.plot.kde()
```
