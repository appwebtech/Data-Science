## Categorical Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Comparing male vs female bills

sns.set_style('darkgrid')
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np 
```

```python
# Adding an estimator aggregate fn 

sns.set_style('darkgrid')
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

```python
# Count plot 

sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
# Box plots

sns.set_style('darkgrid')
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# Specifying a hue
# Will enable you to compare day by day distributions of smokers and non smokers 

sns.set_style('darkgrid')
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
# Violin plots
# Reveals more info on the data but hard to read where the data compares

sns.set_style('darkgrid')
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# Filter by sex 
sns.set_style('darkgrid')
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
# Strip plot 
sns.set_style('darkgrid')
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# Add jitter argument if the density is not well displayed. 
# In my case the distribution is well visible (using a Macbook retina display).
# If sending data like this pass in the 'jitter' arg with a bool of 'True' as people have different display screens.
# 'split' arg has been renamed to 'dodge'. It may be deprecate in future if they update their dependancies.

sns.set_style('darkgrid')
sns.stripplot(x='day',y='total_bill',data=tips, jitter=True, hue='sex',split=True)
```

```python
# Swarm plot 
# Has better representation than a scatter plot

sns.set_style('darkgrid')
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# It's possible to combine a swarm plot with a violin plot 
# I would honestly baptise this as the pomegranate plot. 

sns.set_style('darkgrid')
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# Just poking things on the 'pomegranate' plot 
# You can easily see the KDE plot or curves 
# Personally I dont think this would be ideal for presentation but for exploratory data analysis. 

sns.set_style('darkgrid')
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
# Factor plot 
# It takes a 'kind' arg
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
# Warning: Factor plot was renamed to catplot. Ignore the rubbish above.
# It takes a 'kind' arg

sns.set_style('darkgrid')
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```
