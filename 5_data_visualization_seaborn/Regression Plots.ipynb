## Regression Plots

```python
import seaborn as sns 
%matplotlib inline 
tips = sns.load_dataset('tips')
tips.head()
```

```python
# We want to do an lm plot 
# We get a scatter plot with a linear fit in it. 

sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
# Add more arg's like 'hue'
# We get 2 scatter plots with 2 linear fits. Male and Female have the same linear fit as regards total bill.

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
# We can pass more parms like 'markers' which come from the matplot lib marker symbols 

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
# To edit the size of the markers, we can can pass in a dictionary with as 'scatter_kws'

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'],
          scatter_kws={'s':70})
```

```python
# Do two plots separated by sex category. (Male & Female)

sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
# We can continue adding more args based on what we want to visualize, eg time (Lunch and Dinner), day, hue etc

sns.lmplot(x='total_bill',y='tip',data=tips,row='time', col='day', hue='sex')
```

```python
# Let's change the aspect ration just so we view clearly. {aspect and size}

sns.lmplot(x='total_bill',y='tip',data=tips,row='time', col='day', hue='sex', aspect=0.6, size=8)
```
