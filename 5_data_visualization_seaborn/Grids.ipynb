## Grids

We'll be using Seaborns grids capability to automate sub-plots based on features of our data. 

```python
# iris is a dataset that has measurements of a bunch of different flowers or irises. 

import seaborn as sns 
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Species
iris['species'].unique()
```

```python
# A simple pair plot 

sns.set_style('darkgrid')
sns.pairplot(iris)
```

```python
# pair-grid
# Nothing is displayed.I'll set that to a var (g) and with it I'll map a plot type. 

sns.set_style('darkgrid')
sns.PairGrid(iris)
```

```python
# At first my plot didn't work, an error was thrown of 'plt' not imported. 
# I installed  matplotlib using 'conda' and not 'pip' and all the dependencies should have been fetched. 

import matplotlib.pyplot as plt
# sns.PairGrid(iris)
g = sns.PairGrid(iris) 
g.map(plt.scatter)

sns.set_style('darkgrid')
```

```python
# Specifying what I want on the diagonal plot 

g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

sns.set_style('darkgrid')
```

```python
# Facet grids 

tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Will show an empty grid 

sns.set_style('darkgrid')

g = sns.FacetGrid(tips,col='time',row='smoker')
```

```python
# We need to call map and pass in the kind of plot we want. 

sns.set_style('darkgrid')

g = sns.FacetGrid(tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
# We can pass in some more arg eg tip vs total bill 

sns.set_style('darkgrid')

g = sns.FacetGrid(tips,col='time',row='smoker')
g.map(plt.scatter,'total_bill', 'tip')
```
