## Matrix Plots 

We'll see two forms of matrix plots. 

* Heat maps 
* Cluster maps


```python
import seaborn as sns 
%matplotlib inline 
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
tips.head()

```

```python
# N° of passengers that flew in a given month in a given year

flights.head()
```

```python
# Matrix form needed to call heat map
tc = tips.corr()
```

```python
# Pass in tc
# Colors all values in a gradient using a color scale

sns.set_style('darkgrid')
sns.heatmap(tc)
```

```python
# Anotating the actual numerical values 

sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

```python
flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
# Trying to visualize the data. Save the above data in a variable 'fp' (flight pivot),
# then pass it to the heat map below. 

fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
# As the years advance, more people travel by air. The visibility on the color gradient is amazing. 
sns.heatmap(fp)
```

```python
# Passing in some more args 

sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)

# The second matrix plot in seaborn is the cluster map. 
# Data is represented together with hierarchical clusters. 

sns.clustermap(fp)
```

```python
# Still, we can pass in args to change gradient and enhance view. 

sns.clustermap(fp, cmap='coolwarm')
```

```python
# To normalize the scale we can pass in another argument. In our case the scale will run 
# b2n 0.0 and 1.0 (float's)

sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```
