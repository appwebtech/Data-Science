## Intro and deep dive to Matplotlib 

* Matplotlib is the most popular plotting library for Python 
* It gives you control over every aspect of a figure 
* It was designed to have a similar feel to MatLab's graphical plotting
* You'll need to install it with pip or conda in the terminal 

* **conda install maplotlib** or
* **pip install matplotlib**

* To view some plots that matlib can do: https://matplotlib.org/gallery

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
# There are two ways to create plots:
# 1. Functional method
# 2. Object Oriented method
```

```python
# 1. FUNCTIONAL 
plt.plot(x,y)
```

```python
# Adding x, y label 
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Some Title')
```

```python
# Multi plots
plt.subplot(1,2,1) # No of rows, No of col & plot n° (1st in this case)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# 2. Using Object Oriented API method which is reccommended
# You create figure objects and call methods on them 

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])
```

```python
# Next Plot inside the axes

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
```

```python
# Add x and y lables 

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Some Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4,0.3]) # starts 20% left, 50% bottom, 0.4 & 0.3 position
```

```python
# Plotting inside the axes 

fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LARGER PLOT')
axes2.set_title('SMALLER PLOT')
```

```python
# Some automatic sub-plots

fig,axes = plt.subplots()

axes.plot(x,y)
```

```python
# Adding args in sub-plots

fig,axes = plt.subplots(nrows=1,ncols=2)

# axes.plot(x,y)
```

```python
# Poking things around we can create multiple rows and cols

fig,axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout() # Avoid overlapping
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
```

```python
axes # Its an array and so we can iterate over it
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for joe in axes:
    joe.plot(x,y)

axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[0].set_xlabel('X Label')
axes[0].set_ylabel('Y Label')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
axes[1].set_xlabel('X Label')
axes[1].set_ylabel('Y Label')

plt.tight_layout()
```

### Figure Size and DPI

It's possible to control each aspect of your plot 

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Supposing we want a wide figure (8 x 2 inches)
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Manipulating subplots 

fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
# Manipulating subplots 
# Specify some rows

fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Legend 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2)
ax.plot(x,x**3)
```

```python
# Adding a legend
# Legends are useful for detailing our plots 


fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Adding args for legend
ax.plot(x,x**2,label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend()
```
