### Figure Size and DPI

It's possible to control each aspect of your plot 

```python
fig = plt.figure(figsize=(3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Supposing we want a wide figure (8 x 2 inches)
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# Manipulating subplots 

fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
# Manipulating subplots 
# Specify some rows

fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

### Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# Add color param. It cal also accept hexadecimals eg #FF8C00
# Line width. (pass linewidth or 'lw' as arg)
# Customise line transparency by passing arg 'alpha'
# For dashed line, use linestyle (ls) as arg and pass either '--' 
# for a dashed, '-.' for a dotted or 'steps' for zigzag

ax.plot(x,y,color='orange',linewidth=4, alpha=0.5, linestyle='-.')
```

```python
# Markers 
```

```python
# We know that 'x' is an array of 11 points
# Supposing we want to mark those points on a graph, 
# we can add 'marker' as an arg and pass a string

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='orange',lw=2, marker='+', markersize=10)
# Other markers are: +, *, 1, etc. See documentation
# Use markersize as an arg and pass an int to increase the marker intensity.
```

```python
# You can also specify the marker face color and change the markeredge width

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=1, marker='o', markersize=10,
       markerfacecolor='yellow',markeredgewidth=1.5)
```

```python
# More lines and marker styles 

fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+1, color="red", linewidth=0.25)
ax.plot(x, x+2, color="red", linewidth=0.50)
ax.plot(x, x+3, color="red", linewidth=1.00)
ax.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options â-â, âââ, â-.â, â:â, âstepsâ
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(x, x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

### Plot range

We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
